# シュミレーションと実際の奇跡の違いを比較する

シュミレーションの軌跡が青(kaiseki), 実際の軌跡が赤(practice)

```python
# -*- coding: utf-8
import numpy as np
import matplotlib.pyplot as plt

def main():
    # CSVのロード
    kai_data = np.genfromtxt("kaiseki.csv", delimiter=",", dtype='float')
    pra_data = np.genfromtxt("practice.csv", delimiter=",", dtype='float')

    # x座標, y座標取得
    kai_x = kai_data[:,0]
    kai_y = kai_data[:,1]
    pra_x = pra_data[:,0]
    pra_y = pra_data[:,1]

    # グラフにプロット
    plt.plot((kai_x-kai_x[0])/(kai_x[-1]-kai_x[0])*100, (-kai_y+kai_y[0])/(kai_x[-1]-kai_x[0])*100, label='kaiseki', color='b')
    plt.plot((pra_x-pra_x[0])/(pra_x[-1]-pra_x[0])*100, (-pra_y+pra_y[0])/(pra_x[-1]-pra_x[0])*100, label='practice', color='r')
    g = plt.subplot()
    plt.legend(loc='upper right', fontsize=10)       # 凡例表示
    g.set_aspect('equal', adjustable='box')  # 軸の間隔
    plt.xlabel('x')     # x軸ラベル
    plt.ylabel('y')    # y軸ラベル
    plt.grid()         # グリッド表示
    plt.show()


if __name__ == "__main__":
    main()
```
