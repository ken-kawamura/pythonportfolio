# ライブラリーのインポート

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D, ZeroPadding2D, Dropout,Dense, Flatten, Activation, BatchNormalization
```

# データのインポート

```python
train_data = pd.read_csv('training.csv')
test_data = pd.read_csv('test.csv')
lookid_data = pd.read_csv('IdLookupTable.csv')

print(train_data.shape)
print(test_data.shape)
print(train_data.isna().sum())
```

# データの前処理

```python
train_data.fillna(method='ffill', inplace=True)
```

```python
train_data.head().T
```

```python
test_data.head()
```

```python
image = []
for i in range(0, len(train_data)):
    img = train_data['Image'][i].split(' ')
    img = ['0' if x == '' else x for x in img]
    image.append(img)
    
test_image = []
for i in range(0, len(test_data)):
    test_img = test_data['Image'][i].split(' ')
    test_img = ['0' if x == '' else x for x in img]
    test_image.append(img)
```

```python
image_list = np.array(image, dtype='float')
X_train = image_list.reshape(-1, 96, 96, 1)
test_image_list = np.array(test_image, dtype='float')
X_test = test_image_list.reshape(-1, 96, 96, 1)
```

```python
plt.imshow(X_train[0])
```

```python
training = train_data.drop('Image', axis=1)
training.head().T
```

```python
y_train = []

for i in range(0, len(train_data)):
    y = training.iloc[i, :]
    y_train.append(y)
    
y_train = np.array(y_train, dtype='float')
```

# 訓練データと検証データに分割

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.1, random_state=0)
```

# CNNモデルの構築

```python
model = Sequential()

# Convolution Layer 1
model.add(Conv2D(32, (3, 3), input_shape=(96,96,1))) 
model.add(BatchNormalization(axis=-1))
convLayer01 = Activation('relu')
model.add(convLayer01)

# Convolution Layer 2
model.add(Conv2D(32, (3, 3)))
model.add(BatchNormalization(axis=-1))
model.add(Activation('relu'))
convLayer02 = MaxPooling2D(pool_size=(2,2))
model.add(convLayer02)

# Convolution Layer 3
model.add(Conv2D(64,(3, 3)))
model.add(BatchNormalization(axis=-1))
convLayer03 = Activation('relu')
model.add(convLayer03)

# Convolution Layer 4
model.add(Conv2D(64, (3, 3)))
model.add(BatchNormalization(axis=-1))
model.add(Activation('relu'))
convLayer04 = MaxPooling2D(pool_size=(2,2))
model.add(convLayer04)

model.add(Flatten())

# Fully Connected Layer 5
model.add(Dense(512))
model.add(BatchNormalization())
model.add(Activation('relu'))

# Fully Connected Layer 6                       
model.add(Dropout(0.2))
model.add(Dense(30))

model.summary()
```

# モデルのコンパイル 

```python
model.compile(optimizer='adam', loss='mean_squared_error',metrics=['mae'])
```

# モデルの訓練

```python
model.fit(X_train, y_train, epochs=10, validation_data = (X_val,y_val), batch_size=256, validation_split = 0.2)
```

# モデルの評価

```python
score = model.evaluate(X_val, y_val)
print('Val accuracy:', score[1])
```

# 結果の予測

```python
pred = model.predict(X_test)
pred_list = list(pred)
```

# 提出

```python
rowid = list(lookid_data['RowId'])
imageid = list(lookid_data['ImageId']-1)
featurename = list(lookid_data['FeatureName'])
```

```python
feature = []
for i in featurename:
    feature.append(featurename.index(i))

preded = []
for x, y in zip(imageid, feature):
    preded.append(pred_list[x][y])
```

```python
rowid = pd.Series(rowid, name='RowId')
location = pd.Series(preded, name='Location')

submission = pd.concat([rowid, location], axis=1)
submission.to_csv('submission_keras.csv', index=False)
```
